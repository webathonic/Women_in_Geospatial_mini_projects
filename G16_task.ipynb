```python
"""Number guessing game: Write a Python code that asks a number from the user
and compare it to a number you saved in memory.
If the number added by the user is lower than the hidden number,
print Too low, else print Too high and if correct, print Yay!"""

import random

hidden_number = random.randint(1, 100) # The number the user needs to guess

print("Welcome to the Number Guessing Game!")
print("I have picked a number between 1 and 100. Try to guess it!")

while True:
    try:
        user_guess = int(input("Enter your guess: "))

        if user_guess < hidden_number:
            print("Too low")
        elif user_guess > hidden_number:
            print("Too high")
        else:
            print("Yay! You guessed the number!")
            break # Exit the loop if the guess is correct
    except ValueError:
        print("Invalid input. Please enter an integer.")

```

```python
#Even or Odd number: Ask the user for a number and print
# whether it is even or odd.

try:
    number = int(input("Enter a number: "))
    if number % 2 == 0:
        print(f"The number {number} is even.")
    else:
        print(f"The number {number} is odd.")
except ValueError:
    print("Invalid input. Please enter an integer.")
```

```python
"""Mini calculator: Ask for two numbers and an operator (+, -, *, /)
 and print the result.
 Handle division by zero to appear as an error in a message
 like this: Division by 0 is impossible"""
def mini_calculator():
    try:
        num1 = float(input("Enter the first number: "))
        operator = input("Enter an operator (+, -, *, /): ")
        num2 = float(input("Enter the second number: "))


        result = None

        if operator == '+':
            result = num1 + num2
        elif operator == '-':
            result = num1 - num2
        elif operator == '*':
            result = num1 * num2
        elif operator == '/':
            if num2 == 0:
                print("Division by 0 is impossible.")
                return
            else:
                result = num1 / num2
        else:
            print("Invalid operator. Please use +, -, *, or /.")
            return

        print(f"The result is: {result}")

    except ValueError:
        print("Invalid input. Please enter valid numbers.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

mini_calculator()




```

```python
# Grade printer: Ask a user for a score and print grade:
# 90-100 → A, 80-89 → B, 70-79 → C, 50-69 → D, and 0-49 → E

try:
    score = int(input("Enter your score: "))

    if 90 <= score <= 100:
        print("Your grade is A")
    elif 80 <= score <= 89:
        print("Your grade is B")
    elif 70 <= score <= 79:
        print("Your grade is C")
    elif 50 <= score <= 69:
        print("Your grade is D")
    elif 0 <= score <= 49:
        print("Your grade is E")
    else:
        print("Invalid score. Please enter a score between 0 and 100.")
except ValueError:
    print("Invalid input. Please enter an integer for the score.")
```

```python
# Temperature converter: Ask the user for an input of temperature in Celsius and convert it into Fahrenheit and vice versa.
# You can ask the user to choose the unit first. Depending on the user’s choice, you can do the conversion and print a message on the screen.

def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

def fahrenheit_to_celsius(fahrenheit):
    return (fahrenheit - 32) * 5/9

print("Temperature Converter")
print("1. Celsius to Fahrenheit")
print("2. Fahrenheit to Celsius")

try:
    choice = int(input("Enter your choice (1 or 2): "))

    if choice == 1:
        celsius = float(input("Enter temperature in Celsius: "))
        fahrenheit = celsius_to_fahrenheit(celsius)
        print(f"{celsius}°C is equal to {fahrenheit:.2f}°F")
    elif choice == 2:
        fahrenheit = float(input("Enter temperature in Fahrenheit: "))
        celsius = fahrenheit_to_celsius(fahrenheit)
        print(f"{fahrenheit}°F is equal to {celsius:.2f}°C")
    else:
        print("Invalid choice. Please enter 1 or 2.")
except ValueError:
    print("Invalid input. Please enter a valid number for temperature or choice.")
```

```python
# Weather category classifier: Ask the user for temperature and print a message
# following the conditions: < 20 → Cold, 20-30 → Moderate > 30 → Hot

try:
    temperature = float(input("Enter the temperature in celsius: "))

    if temperature < 20:
        print("Cold")
    elif 20 <= temperature <= 30:
        print("Moderate")
    else: # temperature > 30
        print("Hot")
except ValueError:
    print("Invalid input. Please enter a valid number for the temperature.")
```

```python
# Temperature Analysis: Given a list of daily temperatures (you can define your list with random numbers),
# count how many days were above 35°C. If more than 3 days, print: Heatwave alert.

import random

# Define a list of daily temperatures (you can modify this or use random numbers)
# For demonstration, let's create a list of 10 random temperatures between 25 and 45°C

daily_temperatures = [random.randint(25, 45) for _ in range(10)]

print(f"Daily Temperatures: {daily_temperatures}")

hot_days_count = 0
for temp in daily_temperatures:
    if temp > 35:
        hot_days_count += 1

print(f"Number of days above 35°C: {hot_days_count}")

if hot_days_count > 3:
    print("Heatwave alert.")
```

```python
# Sea level rise projections: Assume sea level rises 3 mm per year.
# Ask the user how many years into the future.
# Calculate and print the total sea level rise according to the number of years specified.

SEA_LEVEL_RISE_PER_YEAR_MM = 3

try:
    years = int(input("Enter the number of years into the future: "))

    if years < 0:
        print("Number of years cannot be negative.")
    else:
        total_sea_level_rise = years * SEA_LEVEL_RISE_PER_YEAR_MM
        print(f"In {years} years, the total sea level rise will be {total_sea_level_rise} mm.")
except ValueError:
    print("Invalid input. Please enter a whole number for the number of years.")
```

```python
# Solar Energy estimator
# Assume that 1 solar panel produces 1.5 kWh per day

try:
    panels = int(input("How many solar panels do you have? "))

    if panels < 0:
        print("Number of panels cannot be negative.")
    else:
        kwh_per_panel_per_day = 1.5
        daily_energy = panels * kwh_per_panel_per_day
        monthly_energy = daily_energy * 30  # Assuming an average of 30 days per month

        print(f"Daily energy production: {daily_energy:.2f} kWh")
        print(f"Monthly energy production: {monthly_energy:.2f} kWh")
except ValueError:
    print("Invalid input. Please enter a whole number for the number of panels.")
```
